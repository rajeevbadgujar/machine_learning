```python
census_data = pd.read_excel('./data/EDA_census.xlsx', header=4, usecols='B:O', skiprows=lambda x: x in range(0,2))
```

```python
census_data = census_data.rename(columns={'Unnamed: 3':'State','Unnamed: 1':'State Code','Unnamed: 2':'Distt. Code',
                                          'Unnamed: 4':'Demography', 'Unnamed: 5':'Age Groups', 
                                          'Unnamed: 6':'Total Persons','Unnamed: 7':'Males','Unnamed: 8':'Females',
                                          'Unnamed: 9':'Illiterate Total', 'Unnamed: 10':'Illiterate Males',
                                          'Unnamed: 13':'Literate Males','Unnamed: 14':'Literate Females',
                                          'Unnamed: 12':'Literate Total', 'Unnamed: 11':'Illiterate Females'})
census_data.head()
```

```python
age_groupby = census_data.groupby(['Age Groups'])

age_groupby.mean()['Illiterate Females']/age_groupby.mean()['Females']
```

```python
age_groupby.mean()['Literate Total']/age_groupby.mean()['Total Persons']
```
