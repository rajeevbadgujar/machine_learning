# Step 2: Data preparation

```python
# Create dummy columns for month
bike = pd.get_dummies(data=bike, prefix='month', columns=['mnth'], prefix_sep='_', drop_first=True)
print(bike.shape)
bike.head()
```

Lets create a separate dataframe with only the columns which we will use to build model.
From our earlier observations we can drop temp as atemp and temp have very strong correlation.

```python
df_bike = bike[['season','yr','holiday','weekday','workingday','weathersit','temp', 'atemp','hum','windspeed', 'month_2', 
                'month_3', 'month_4', 'month_5', 'month_6', 'month_7', 'month_8', 'month_9', 'month_10', 'month_11',
                'month_12','cnt']]
print(df_bike.shape)
df_bike.head()
```

```python
# Train and test split
bike_train, bike_test = train_test_split(df_bike, train_size=0.7, random_state=100)
```

```python
print(bike_train.shape)
print(bike_test.shape)
```

```python
bike_train.head()
```

```python
bike_test.head()
```

```python
# Lets describe numerical variables in train data
bike_train.describe()
```

We have to scale following variables 
1. season
2. weekday
3. weathersit
4. atemp
5. hum
6. windspeed

The values of variables 
1. year
2. holiday
3. workingday
4. months 
are already 0 or 1 so we need not scale these variables.

**Note: we will use MinMaxScaler first and then we will build a model by standardising the numerical variables**

```python
# MinMax scaler
scaler = MinMaxScaler()
```

```python
num_var = ['season','weekday','weathersit', 'temp', 'atemp','hum','windspeed', 'cnt']

bike_train[num_var] = scaler.fit_transform(bike_train[num_var])

bike_train.head()
```

```python
bike_train.shape
```

```python
bike_train.describe()
```

```python
# Divide the independent and dependent variables
y_train = bike_train.pop('cnt')
X_train = bike_train
```

# Step 3: Model Building

```python
# Add a constant
X_train_lm = sm.add_constant(X_train)

# Create a first fitted model
lr = sm.OLS(y_train, X_train_lm).fit()
```

```python
# The learned parameters
lr.params
```

```python
lr.summary()
```

## Observations
- the p-value for month_7, month_11, month_12 and month_8 are high
- the p-value for atemp is high as well

Lets calculate VIF for all the features and then we can drop features which are insignificant.

## Check for VIF

```python
vif = pd.DataFrame()
vif['features'] = X_train.columns
vif['VIF'] = [variance_inflation_factor(X_train.values, i) for i in range(X_train.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by='VIF', ascending=False)
vif
```

## Observations
- The VIF for atemp and temp are very high
- We will drop these variable and recompute the VIF

```python
# drop atemp
X_train_1 = X_train.drop(['temp', 'atemp'], axis=1)

# rebuild the model with remaining variables
X_train_lm = sm.add_constant(X_train_1)

lr = sm.OLS(y_train, X_train_lm).fit()
```

```python
print(lr.params)
print(lr.summary())
```

```python
# recalculate the VIF after dropping temp and atemp
vif = pd.DataFrame()
vif['features'] = X_train_1.columns
vif['VIF'] = [variance_inflation_factor(X_train_1.values, i) for i in range(X_train_1.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by='VIF', ascending=False)
vif
```

## Observations
- The p-value of season is zero however the VIF is high 
- the p-value and vif for hum is high

We will drop season

```python
# drop season
X_train_2 = X_train_1.drop(['season'], axis=1)

# rebuild the model with remaining variables
X_train_lm = sm.add_constant(X_train_2)

lr = sm.OLS(y_train, X_train_lm).fit()
```

```python
print(lr.params)
print(lr.summary())
```

```python
# recalculate the VIF after dropping season
vif = pd.DataFrame()
vif['features'] = X_train_2.columns
vif['VIF'] = [variance_inflation_factor(X_train_2.values, i) for i in range(X_train_2.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by='VIF', ascending=False)
vif
```

```python
# Now lets drop hum and rebuild the model and calculate VIF
# drop hum
X_train_3 = X_train_2.drop(['hum'], axis=1)

# rebuild the model with remaining variables
X_train_lm = sm.add_constant(X_train_3)

lr = sm.OLS(y_train, X_train_lm).fit()

# recalculate the VIF after dropping season
vif = pd.DataFrame()
vif['features'] = X_train_3.columns
vif['VIF'] = [variance_inflation_factor(X_train_3.values, i) for i in range(X_train_3.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by='VIF', ascending=False)
```

```python
print(lr.params)
print(lr.summary())
```

```python
vif
```

## Observations
- The p-value for holiday is high however the VIF is low 
- The p-value for workingday is less than 0.05 and VIF is low too.

As these two columns represent the same information, dropping any one of them should not affect our model accuracy. Lets drop the column with high p-value i.e. holiday

```python
# Now lets drop holiday and rebuild the model and calculate VIF
# drop holiday
X_train_4 = X_train_3.drop(['holiday'], axis=1)

# rebuild the model with remaining variables
X_train_lm = sm.add_constant(X_train_4)

lr = sm.OLS(y_train, X_train_lm).fit()

# recalculate the VIF after dropping holiday
vif = pd.DataFrame()
vif['features'] = X_train_4.columns
vif['VIF'] = [variance_inflation_factor(X_train_4.values, i) for i in range(X_train_4.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by='VIF', ascending=False)
```

```python
print(lr.params)
print(lr.summary())
```

```python
vif
```

The VIFs and p-values both are within an acceptable range. So we go ahead and make our predictions using this model only.

## Observations
- we dropped variables **temp, atemp, hum, season and holiday**
- **Months, windspeed and weather situation** have higher impact on the bike reservations
- There is positive impact on the bike reservations over base year of 2018

# Step 4: Residual Analysis of the train data

```python
# now our X_train is X_train_4
y_train_pred = lr.predict(X_train_lm)
```

```python
# plot histogram of error between y_train and predicted values of y on training data
plt.figure()
sns.displot(data=(y_train - y_train_pred), bins=20)
plt.title('Error Terms', fontsize = 20)     
plt.xlabel('Errors', fontsize = 18) 
plt.show()
```

## Observations
- The error terms are centered around Zero
- The error terms show normal distribution

# Step 5: Predictions

```python
# scale the variables using the values fit on training set
bike_test[num_var] = scaler.transform(bike_test[num_var])

# Use only the features that we used to build the model
bike_test = bike_test.drop(['season', 'atemp', 'temp', 'hum', 'holiday'], axis=1)

# divide into X_test and y_test 
y_test = bike_test.pop('cnt')
X_test = bike_test

# Adding constant variable to test dataframe
X_test = sm.add_constant(X_test)

# predictions
y_pred = lr.predict(X_test)
```

```python
plt.figure()
sns.scatterplot(x=y_test, y=y_pred)
plt.title('y_test vs y_pred')
plt.xlabel('y_test')
plt.ylabel('y_pred')
```

```python
r2_score(y_test, y_pred)
```
