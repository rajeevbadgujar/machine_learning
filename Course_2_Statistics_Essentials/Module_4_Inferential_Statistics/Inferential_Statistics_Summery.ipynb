# Inferential Statistics
Its not always possible to check all the data because of sheer volume of it. So we work with smaller sample of data and try to infer insights from it. This process of **inferring** insights from sample data is called **Inferential Statistics**.


# Basics of Probability




### Definitions

**Experiment:** a repeatable procedure with a set of possible results. e.g. throwing dice

**Outcome**: a possible result of an experiment. e.g. dice showing 6

**Sample Space**: all possible outcomes of an experiment. e.g. dice can show 1,2,3,4,5,6

**Sample Point**: just one of the possible outcomes of an experiment. e.g. dice showing 6

**Event**: one or more outcomes of an experiment. e.g. dice rolling a 5 is just one outcome whereas dice rolling an even number is more than one outcome - **a number and its even**

**Relative Frequency**: how often something happens divided by all outcomes. e.g. India won 10 matches out of 15 so the frequency is 10 and relative frequency is 2/3.

**Random Variable**: it measures the outcome of the experiment. e.g. number of Red balls drawn by player. Generally denoted by X.

**Discrete Random Variable**: The random variable which can take a distinct value. The value can be very small or very large but its finite. e.g. player winning 100 if dice rolls a 6 else looses 20 or number children born in a day in India

**Continuous Random Variable**: The random variable which can take any value in an interval. 

**Probability Distribution**: The distribution of all possible values for our Random Variable. It can be a table, an expression or chart.

**Expected Value**: The value of X we expect after running the experiment once. The experiment here may constitue as number of repeatition. It is also called the **expectation, average, and mean value**. So for a Random Variable X which can takes values x1, x2,...,xn, the Expected Value is

### Addition Rule of Probability
The addition rule states that the probability two events is the sum of the probability that either one of them will happen minus the probability that both will happen.

        P(A or B) = P(A) + P(B) - P(AB)     OR
        P(A ∪ B) = P(A) + P(B) − P(A ∩ B)
e.g. the probability of rolling a 3 or 4 on dice is sum of probability of rolling 3 and 4. Here as dice can show only one number and not both, the probability of rolling both numbers is 0.
         P(roll 3 or 4) = 1/6 + 1/6
Now the probability of rolling a 4 or even number will be sum of probability of rolling 4 or even number minus probabilty of rolling a 4 because 4 is an even number.
         P(roll 4 or even) = 1/6 + 1/2 - 1/6

### Multiplication Rule of Probability
The multiplication rule states that the probability that A and B both occur is equal to the probability that A occurs times the conditional probability that B occurs provided A occurs.
Note that here the events A and B are not independent and the experiments are without replacemenet.

        P(A ∩ B) = P(A) ⋅ P(B|A)
        the term P(B|A) indicates the probability of B occuring provided A occurs.

Consider following example:

Event A = getting an ace from a pack of cards\
Event B = getting an ace from a pack of cards 

The probability of A is 4/52 and that of B is 3/51 because we already have drwan an ace. So as per the multiplication rule the probability of A and B is 1/13 * 1/17.

**Independent event**
Two events are said to be independent if occurance of one event does not affect the probability of other. So our earlier equation of multiplication rule becomes 
    
        P(A ∩ B) = P(A) ⋅ P(B)
        here P(B|A) = P(B) because probability of A and B are independent.
        
Taking the same exaple of drawing aces. Here we can say event A and B are independent if we place the ace drawn in first attemp back in the pack. Thus the probability of A does not affect the probabilty of B. 
Hence by the multiplication rule the probaility of A and B is 4/52 * 4/52 i.e. 1/13 * 1/13.

# Continuous Probability Distribution
The probability distribution for a continuous random variable X. A continuous random variable X can take **infinite** values. Therefore the probability of X taking on any **one specific value is zero**. So we often talk in terms of range of values X can assume. The probability we will be tackling here would be **cumulative probability.** The probability of X falling between two values a and b equals the integral (area under the curve) from a to b.


### Cumulative Disribution Function (CDF)
Monotonically non-decreasing function. Plots cumulative probability of X against X.


Note that the CDF for discrete variable is a bar chart because in case of discrete variable we are interested in probability of a discrete value say 0, 1, 2, 3 and the probability does not change for value between 1 and 1.99. 


### Probability Density Function (PDF)

### Sampling Distributions
From the population (N) multiple small samples of size (n) are taken and sampling mean is computed for each. The distribution of these sampling means is called as sampling distribution. 
The sampling mean is best estimate for our population mean because taking a large number of samples removes the sampling bias by accomodating the diversity inherent in the population. So larger the number of samples, closer the sampling mean gets to population mean. Though in practice its not possible to take a very large number of samples, therefore we often give range estimates for population mean. 

#### Sampling Distribution Standard Deviation 
This gives the range where population mean is most likely to fall. Also known as **Standard Error (S.E.)**, because its the error in infering the population mean. 



### Central Limit Theorem
The central limit theorem says that, for any kind of data, provided a high number of samples has been taken, the following properties hold true:

1. Sampling distribution’s mean (μ¯X) = Population mean (μ)
2. Sampling distribution’s standard deviation (Standard error) = σ/√n
3. For n > 30, the sampling distribution becomes a normal distribution

So using the Standard Error, we can get the very good estimate of the probability of range in which the population mean would lie. 

Probability associated with the cliam is called **confidence level**, denoted in %.\
The maximum error made in sample mean is called **margin of error** which says how far our population mean is from actual mean. (2 S.E. or 1.65 S.E. etc)\
The interval of values for our population mean is called **confidence interval** (μ - S.E, μ + S.E.)

# Sampling methods 
## Random Sampling 
- randomly selected from population 
- each sample has equal chance of selection
- advantage is that its unbiased if the population size is large, but if population size is not large, it will result in biased sample
- disadvantage is that its time consuming given that the population size is large

## Stratified Sampling 
- the population is divided into segments 
- samples are selected from each segment rnadomly such that each segment has same representaion as in original population
- unbiased 
- time consuming 

## Volunteer Sampling
- volunteers for any survey form the sample
- may be biased 

## Opportunity Sampling
- people around the surveyor form their sample space 
- unrepresentative of the actual sample 
- biased by reseracher
- quick, convenient and economical
