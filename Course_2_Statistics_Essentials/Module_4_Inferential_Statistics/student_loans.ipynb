```python
import numpy as np
import pandas as pd
```

```python
student_loan_data = pd.read_csv('./data/Inferential Statistics - Student Loan.csv')
```

```python
student_loan_data['Recovery (%)'] = student_loan_data['Recovery (%)'].apply(lambda x: float(x[0:-1])/100)
```

```python
student_loan_data['losses'] = student_loan_data.apply(lambda row: (1-row['Recovery (%)'])*row['Exposure at Default (in lakh Rs.)']*row['Probability of Default'], axis=1)
```
