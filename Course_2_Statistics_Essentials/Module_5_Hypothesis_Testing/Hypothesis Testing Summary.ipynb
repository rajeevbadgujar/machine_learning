# Hypothesis Testing
Hypothesis is a claim or an assumption that you make about one or more population parameters.\
Hypothesis Testing is to confirm our conclusion (hypothesis) about population parameters (which are known from EDA). Through Hypothesis testing we can determine if there is enough evidence to conclude that the hypothesis about the population parameter is true or not.

## Types of Hypothesis 

#### Null Hypothesis (H0)
The status quo. Null hypothesis is never accepted. If enough evidence is available to reject the null hypothesis, the null hypothesis is rejected. Similarly in case of not enough evidence, it does not mean that we accept the null hypothesis it just means that there is not enough evidence to support the alternate hypothesis i.e. we fail to reject the null hypothesis in favor of null hypothesis.
If your claim statement has words like **“at least”, “at most”, “less than”, or “greater than”**, you cannot formulate the null hypothesis just from the claim statement (because it’s not necessary that the claim is always about the status quo).\
The null hypothesis always has the following signs:  =  OR   ≤   OR    ≥.

        at least is ≥.
        at most is ≤.


#### Alternate Hypothesis (H1)
The challange to the status quo. Also called as research hypothesis. If sufficient evidence is available to support the alternate hypothesis, we reject the null hypothesis.
The alternate hypothesis always has the following signs:  ≠   OR  >   OR    <.

**Situation 1:** Flipkart claimed that its total valuation in December 2016 was at least 14 billion. Here, the claim contains ≥ sign (i.e. the at least sign), so the null hypothesis is the original claim.\
Total Valuation ≥ 14 Billion --> Null hypothesis\
Total Valuation < 14 Billion --> Alternate hypothesis

**Situation 2:** Flipkart claimed that its total valuation in December 2016 was greater than $14 billion. Here, the claim contains > sign (i.e. the ‘more than’ sign), so the null hypothesis is the complement of the original claim.\
Total Valuation ≤ 14 Billion --> Null hypothesis\
Total Valuation > 14 Billion --> Alternate hypothesis

## Making a Decision

### Critical Value Method

#### Critical Value:
The value at which we can reject the null hypothesis.

**Lower Critical Value:** The lower bound of the critical value.

**Upper Critical Value:** The higher bound of the critical value.

The lower or upper values depend on the type of alternate hypothesis i.e. it is not necessary for all alternate hypothesis to have lower and upper critical value. The alternate hypothesis can be directional or non-directional. 
You can tell the type of the test and the position of the critical region on the basis of the ‘sign’ in the alternate hypothesis.

       ≠ in H₁    →   Two-tailed test       →     Rejection region on both sides of distribution
       < in H₁    →   Lower-tailed test     →     Rejection region on left side of distribution
       > in H₁    →   Upper-tailed test     →     Rejection region on right side of distribution

#### Critical Region
The region below or above the LCV or UCV. If the value falls in critical region we reject the null hypothesis. 

#### Acceptance Region
The region which falls between LCR and UCR where we fail to reject the null hypothesis.

#### Critical Value Method
1. Calculate the value of Zc from the given value of α (significance level)
2. Calculate the critical values (UCV and LCV) from the value of Zc
        UCV = μ + Zc * S.E.
        LCV = μ - Zc * S.E.
3. Make the decision on the basis of the value of the sample mean ¯x with respect to the critical values (UCV AND LCV)

### p value method
1. Calculate Z value for sample mean
2. Calculate probability for sample mean using Z table 
3. 1-probability of sample mean is the p-value
4. Multiply by 2 is its 2-tailed test
5. Reject null hypothesis if p-value is smaller than the alpha
6. Fail to reject null hypothesis if p-value is greater than alpha 

#### Types of Errors


**Type 1 error:**
- Null hypothesis is rejected when its correct
- denoted by α

**Type 2 error:**
- Null hypothesis is not rejected when its incorrect
- denoted by β
- inversely proportional to α



**Note:Both critical value and p value method together are called Z test**
