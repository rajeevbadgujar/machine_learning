```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
prices = pd.read_csv('./data/EDA_Gold_Silver_prices.csv')

prices.head()
```

```python
prices.corr()
```

```python
extract_year = lambda x: x[-2:]

prices['Year'] = prices['Month'].apply(extract_year)
```

```python
extract_year('Feb-98')
```

```python
prices[prices['Year'] == '08'].corr()
```

```python
years = prices['Year'].value_counts().reset_index()
```

```python
years.columns=['year', 'counts']
```

```python
for x in years.year:
    print("\n")
    print(x)
    print(prices[prices['Year'] == x].corr())
```
