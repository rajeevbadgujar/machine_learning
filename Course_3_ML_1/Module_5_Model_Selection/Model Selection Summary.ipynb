# Model Selection

1. How to select best fit model for given data?
2. How to compare different models without building them explicitly?

**_The central issue in machine learning is how to extrapolate learnings from the finite available data to all possible inputs of the same kind?_**

## Occam's razor
- When in dilemma, choose simpler model
- the model should be as simple as necessary but not simpler than that
- simplicity can be in terms of space complexity, time complexity, representation etc 
- simplicity is not expected to cost robustness of the model

### Simpler models vs complex models
- simpler models are usually more generic i.e. more likely to be applicable on unseen data 
- in contrast a complex model is specialized for a set of data which may not give desired results on unseen data 
- simpler models require fewer training samples than complex models 
- simpler models are more robust 
- simpler models may make more errors in training as against complex models which may do very well on the training data but pooerly on unseen data 
- simpler models make fewer assumptions 

## Overfitting
- models are trained on training data but their efficacy is measured on unseen data in real world

## Bias-Variance Tradeoff
- Variance is the variance in output of the model when some training data changes. 
- Complex models have high variance thus needs more changes when training data changes
- Thus variance for a model is degree of change in the model itself with respect to changes in training data.
- Variance is about consistancy
- Bias quantifies how accurate is the model on unseen (test) data.
- Complex model with large enough training data can do very accurate job of prediction.
- Bias is about accuracy

## Practical issues in choosing model
- Type of data 
- Data quality 
- Dimentiality of the data 
- Nature of variables - continuous vs categorical 

## Class of models
Based on the the of data following are the classes of models 
- High dimentional data 
- Noisy data 
- Real time data 
- Large amount of data 
- Missing data 

## Regularization
- to prevent the model from becoming complex 
- its part of the learning algorithm

## Hyperparameters
- input to learning algorithm 
- tuning parameters to keep the model simple 

## Training data 
- used to train the model
- statistical charactaristics of the original data must be preserved

## Validation data 
- used to tune hyperparameters
- statistical charactaristics of the original data must be preserved

## Test data 
- used to test the model
- statistical charactaristics of the original data must be preserved

## Cross-validation
- Split the original data into different subets, maybe shuffle the data 
- use combination of these subsets as training and test data to build different models 
- choose best performing model from built models
